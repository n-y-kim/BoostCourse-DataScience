```python
is_southkorea = df['iso_code'] == 'KOR'

# 조건를 충족하는 데이터를 필터링하여 새로운 변수에 저장합니다.
southkorea = df[is_southkorea]

# 결과를 출력합니다.
southkorea
```

```python
southkorea.plot.area()
```

```python
southkorea.plot.line(x="date", y="people_fully_vaccinated_per_hundred")
```
