```python
import pandas as pd
```

```python
df = pd.DataFrame(pd.read_csv("data/gdp_csv.csv"))
```

```python
df
```

```python
is_southkorea = df['Country Code'] == 'KOR'

# 조건를 충족하는 데이터를 필터링하여 새로운 변수에 저장합니다.
southkorea = df[is_southkorea]

# 결과를 출력합니다.
southkorea
```

```python
southkorea.plot.line(x="Year")
```

```python
is_AUS = df['Country Code'] == 'AUS'

# 조건를 충족하는 데이터를 필터링하여 새로운 변수에 저장합니다.
AUS = df[is_AUS]

# 결과를 출력합니다.
AUS
```

```python
AUS.plot.line(x='Year')
```

```python
is_2016 = df['Year'] == 2016

# 조건를 충족하는 데이터를 필터링하여 새로운 변수에 저장합니다.
year2016 = df[is_2016]

# 결과를 출력합니다.
year2016
```

```python
year2016.plot.line()
```

```python
year2016.plot.scatter(x='Country Name',y='Value', s=10)
```

```python
year2016.plot.bar(x='Country Name',y='Value')
```
